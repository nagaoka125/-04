# **マインスイーパー**

## **概要**

このプログラムは名前の通りマインスイーパーを遊ぶことができるプログラムです。
しかし、旗を立てる機能はゲームのテンポが悪くなったのでありません。
このプログラムでは自分でマインスイーパーの盤面の大きさを決めることができ、難易度を調節することができます。

## **操作方法**

まず、最初に盤面一辺の長さを標準入力で`盤面の一辺の長さを決めてください(4以上)`に4以上の整数を入力します。4以上の整数以外を入力しても再び入力を求められます。盤面のボタンを左クリックするとタイルを開くことができます。あとはマインスイーパーのルールに従ってゲームをプレイしてください。ゲーム終了時に次のゲームを連続でできないのでその都度再実行してください。<br>
**マインスイーパーのルール**<br>
マインスイーパーは盤面に埋まっていいる地雷を避けて安全なマスをすべて開けることができればゲームクリア、地雷が埋まっているマスを開けてしまったらゲームオーバーです。<br>
安全なマスを開くと数字が開いたマスに表示されます。これは数字の数だけ地雷が周囲のマスに埋まっていることを表しています。<br>
例えば、開いたマスの数字が0なら周囲のマスに地雷はないので周囲のマスもすべて安全マスになります。しかし、数字が2なら周囲のマスのうちのどこか2マスに地雷が埋まっています。<br>
つまり、以下の例の場合だと<br>
例：A B C D E <br>
　　F 0 1 G H <br>
　　I Ｊ K Ｌ 2 <br>
まず、0の周りはすべて安全なので、A、B、C、F、I、J、Kは安全マスです。<br>
次に、1のマスが示している地雷は、D、G、Lのうちのどれか1つです。<br>
そして、2が示している地雷は、G、H、Lのうち2つです。<br>
よって、1と2で地雷が埋まっている可能性があるマスが被っているので、2と1の共通の地雷はG、Lのどちらかに埋まっており、Dは安全、2のもう一方の地雷はHに埋まっている。GとLはK、残ったEはDを開けば判明する。<br>
このようにして安全マスの数字をもとに地雷を推測して避けながら、すべての安全マスを開くことでゲームクリアできます。

しかし、今回の長期課題のプログラムは、あまりいいプログラムでないと思うしもっと省略できる部分もあると思うので、自分がまだまだプログラミングを身につけられていないことがはっきりとわかりました。だから、今後はより勉強の時間を増やして改善できるようにするために頑張ります。

## **作成時間**

作成時間：**約11時間10分**



```python
%reset -f
import ipywidgets as widgets #タイルを作れるライブラリ
from IPython.display import display, clear_output #タイルを表示するライブラリ
import random as r

#ゲームの設定----------------------------
print("盤面の一辺の長さを決めてください(4以上)")
length = input()
while True: #無限ループ
  try:
      length = int(length) #int型に変換
      if length >= 4: #4以上ならループ終了
          break
      else:
          clear_output() #非表示
          print("対応していない値が入力されました","値を入力し直してください",sep=",")
          length = input() #再入力
  except ValueError: #入力値がint型以外のとき
      clear_output()
      print("対応していない値が入力されました","整数を入力し直してください",sep=",")
      length = input() #再入力

rows = length      #タイルの行数
columns = length   #タイルの列数
mines =int((rows*columns) / 5)     #地雷の数
tile_list = [0]*(rows*columns-mines) #行*列から地雷の数を引いた数だけ要素を追加
print(f"地雷の数は{mines}個です")

#地雷の配置
mine = r.sample(range(rows*columns),mines)  #被りなしにランダムで地雷を選ぶ
while 0 in mine:  #0が出なくなるまで繰り返す
    print(0)
    mine = r.sample(range(rows*columns),mines)
#print(list(mine))確認用

def click_tile(row, col): #クリックしたときに呼び出す関数
  global other_tile  #グローバル変数化
  other_tile = []
  x = columns-2
#周囲のタイル番号をother_tileに追加する処理--------------------------
  #0列目のとき
  if col == 0:
      for r in range(3):
        x = x+1
  #周りのタイル番号が0<タイル番号<rows*columnsかつ5の倍数でなければ追加
        if (row*columns+col+1-x) % columns != 0 and 0 < row*columns+col+1-x :
            other_tile.append(row*columns+col+1-x)
        if (row*columns+col+1+x) % columns != 0 and row*columns+col+1+x < rows*columns:
            other_tile.append(row*columns+col+1+x)
      other_tile.append(row*columns+col+1+1)

  #最後の列のとき
  elif col == columns-1:
      for r in range(3):
        x = x+1
  #周りのタイル番号が0<タイル番号<rows*columnsかつ5で割った余りが1でなければ追加
        if (row*columns+col+1-x) % columns != 1 and 0 < row*columns+col+1-x :
            other_tile.append(row*columns+col+1-x)
        if (row*columns+col+1+x) % columns != 1 and row*columns+col+1+x <= rows*columns:
            other_tile.append(row*columns+col+1+x)
      other_tile.append(row*columns+col+1-1)

  #0行目と最後の行のとき
  elif row == 0 or row == rows-1:
      other_tile.append(row*columns+col+1+1)
      other_tile.append(row*columns+col+1-1)
      for r in range(3):
        x = x+1
  #周りのタイル番号が0<タイル番号<rows*columnsなら追加
        if row*columns+col+1-x > 0:
            other_tile.append(row*columns+col+1-x)
        if row*columns+col+1+x < rows*columns:
            other_tile.append(row*columns+col+1+x)

 #内側のタイルのとき
  else:
      other_tile.append(row*columns+col+1-1)
      other_tile.append(row*columns+col+1+1)
      for r in range(3):
          x = x+1
          other_tile.append(row*columns+col+1-x)
          other_tile.append(row*columns+col+1+x)

#  print(other_tile)
#ここからクリックしたときの判定--------------------
  if row*columns+col+1 in mine:  #クリックしたタイルが地雷リストにあれば
      clear_output() #タイルをすべて非表示
      print("地雷のタイルを開きました")
      print("---GAMEOVER---")
  else:
      tile_num = []
      for item in  other_tile:
          if item in mine:
               tile_num.append(item)  #地雷と同じ要素がある場合追加
      n = len(tile_num)
      tiles[row][col].description = f'[{str(n)}]'  #クリックしたtileの表示変更
      tile_list.pop(0) #セーフの場合要素を1つ削除
      if len(tile_list) == 0:  #リストの要素がなくなったらゲーム終了
          clear_output()
          print("地雷を回避しました")
          print("---GAME CLEAR!---")
         #print(tile_list)確認用

#タイルの生成-----------------
tiles = []
for row in range(rows):
    tile_row = []
    for col in range(columns):
      tile = widgets.Button(description = ' ',layout = widgets.Layout(width='50px',height='50px'))  #タイルの設定
      tile.on_click(lambda b, r=row, c=col: click_tile(r,c))  #クリックしたときのコールバック関数の選択
      tile_row.append(tile)  #設定したタイルを追加
    all_tile =  widgets.HBox(tile_row) #横にもタイルを追加
    tiles.append(tile_row)
    display(all_tile)  #タイルを表示
```
